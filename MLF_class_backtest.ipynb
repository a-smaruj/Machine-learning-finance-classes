# Strategy

```python
#def buy(self, *, size=.6, limit=None, stop=None, sl=None, tp=None)
#def sell(self, *, size=.6, limit=None, stop=None, sl=None, tp=None)
#limit- nie wiecej/nie mniej niż,
#sl (stop loss) - wartosc kiedy w koncu kupujemy, tp(take profit) - wartosc gdy sprzedajemy, by miec wreszcie zysk
#TRADING: sl śledzi cenę, sl=2%, więc dokupi rośnie wartość sl rośnie, jak spada, to sie zatrzymuje i mogą się wtedy spotkać wartosci
```

```python
class SmaCross(Strategy):
    n1 = 10
    n2 = 20

    def init(self):
        close = self.data.Close
        self.sma1 = self.I(SMA, close, self.n1)
        self.sma2 = self.I(SMA, close, self.n2)

    def next(self):
        if crossover(self.sma1, self.sma2):
            self.position.close()
            self.buy()
        elif crossover(self.sma2, self.sma1):
            self.position.close()
            self.sell()
```

```python
#cash - wartosc kasy, commision - oplata transakcyjna
bt = Backtest(EURUSD, SmaCross, cash=10_000, commission=.002)
stats = bt.run()
print(stats)
bt.plot()
```

```python
#cash - wartosc kasy, commision - oplata transakcyjna
bt = Backtest(GOOG, SmaCross, cash=10_000, commission=.002)
stats = bt.run()
print(stats)
bt.plot()
```

# Strategy - dodanie parametrów sl, tp

```python
class SmaCross_2(Strategy):
    n1 = 10
    n2 = 20

    def init(self):
        close = self.data.Close
        self.sma1 = self.I(SMA, close, self.n1)
        self.sma2 = self.I(SMA, close, self.n2)

    def next(self):
        if crossover(self.sma1, self.sma2):
            self.position.close()
            price=self.data.Close
            long_sl = price * 0.99
            long_tp = price * 1.01
            self.buy(size=0.5, sl = long_sl, tp = long_tp)
        elif crossover(self.sma2, self.sma1):
            self.position.close()
            price=self.data.Close
            short_sl = price * 1.01
            short_tp = price * 0.95
            self.sell(size=0.5, sl = short_sl, tp = short_tp)
```

```python
bt = Backtest(EURUSD, SmaCross_2, cash=10_000, commission=.002)
stats = bt.run()
print(stats)
bt.plot()
```

```python
bt = Backtest(GOOG, SmaCross_2, cash=10_000, commission=.002)
stats = bt.run()
print(stats)
bt.plot()
```

# Strategy - wyjście jeśli short lub long, by nie iść w nieskończoność

```python
class SmaCross_3(Strategy):
    n1 = 10
    n2 = 20

    def init(self):
        close = self.data.Close
        self.sma1 = self.I(SMA, close, self.n1)
        self.sma2 = self.I(SMA, close, self.n2)

    def next(self):
        if crossover(self.sma1, self.sma2):

          price=self.data.Close
          long_sl = price * 0.99
          long_tp = price * 1.01

          if (self.position.is_short):
            self.position.close()
            self.buy(size=0.5, sl = long_sl, tp = long_tp)
          elif not(self.position.is_long or self.position.is_short):
            self.buy(size=0.5, sl = long_sl, tp = long_tp)

        elif crossover(self.sma2, self.sma1):

          price=self.data.Close
          short_sl = price * 1.01
          short_tp = price * 0.95

          if (self.position.is_long):
            self.position.close()
            self.sell(size=0.5, sl = short_sl, tp = short_tp)
          elif not(self.position.is_long or self.position.is_short):
            self.sell(size=0.5, sl = short_sl, tp = short_tp)
```

```python
bt = Backtest(EURUSD, SmaCross_3, cash=10_000, commission=.002)
stats = bt.run()
print(stats)
bt.plot()
```

```python
bt = Backtest(GOOG, SmaCross_3, cash=10_000, commission=.002)
stats = bt.run()
print(stats)
bt.plot()
```

# Optymalizacja strategii

```python
bt = Backtest(EURUSD, SmaCross_3, cash=10_000, commission=.002)
stats = bt.optimize(n1=range(5, 30, 5),
                    n2=range(10, 70, 5),
                    maximize='Equity Final [$]',
                    constraint=lambda param: param.n1 < param.n2)
stats
```

```python
print(stats._strategy)
```
